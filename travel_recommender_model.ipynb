```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.tokenize import RegexpTokenizer

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Define a function for preprocessing
def preprocess_text(text):
    # Lowercasing
    text = text.lower()

    # Tokenization
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(text)

    # Removing stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    # Stemming
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]

    # Join the tokens back into a single string
    preprocessed_text = ' '.join(stemmed_tokens)

    return preprocessed_text

# Apply preprocessing to the 'review' column
df['preprocessed_review'] = df['review'].apply(preprocess_text)
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download NLTK resources (only required once)
nltk.download('vader_lexicon')

# Initialize SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Function to get sentiment scores
def get_sentiment(text):
    # Get polarity scores
    scores = sid.polarity_scores(text)
    # Classify sentiment based on compound score
    if scores['compound'] >= 0.05:
        return 'Positive'
    elif scores['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to the reviews
df['Sentiment'] = df['review'].apply(get_sentiment)
```

```python
# Create an empty DataFrame to store user locations and ratings
user_ratings_locations_df = pd.DataFrame(index=pivot_ratings.index, columns=['locations'])

# Iterate over each row in the pivot_ratings DataFrame
for user_id, row in pivot_ratings.iterrows():
    # Get the locations and ratings where the rating is not NaN
    locations_with_ratings = [(loc, rating) for loc, rating in row.items() if not pd.isnull(rating)]
    # Sort the locations alphabetically
    locations_with_ratings.sort(key=lambda x: x[0])
    # Extract the locations and ratings separately
    locations = [loc for loc, _ in locations_with_ratings]
    ratings = [rating for _, rating in locations_with_ratings]
    # Join the locations into a comma-separated string
    locations_str = ', '.join(locations)
    # Join the ratings into a comma-separated string
    ratings_str = ', '.join(map(str, ratings))
    # Store the locations and ratings for the user in the user_ratings_locations_df DataFrame
    user_ratings_locations_df.at[user_id, 'locations'] = locations_str
    user_ratings_locations_df.at[user_id, 'ratings'] = ratings_str

# Display the DataFrame
user_ratings_locations_df
```

```python
import pandas as pd

# Split the comma-separated locations string into individual binary columns (one-hot encoding)
encoded_locations = user_ratings_locations_df['locations'].str.get_dummies(sep=', ')

# Drop the original 'locations' column
user_ratings_locations_encoded = user_ratings_locations_df.drop(columns=['locations'])

# Concatenate the encoded locations DataFrame with the original DataFrame
user_ratings_locations_encoded = pd.concat([user_ratings_locations_encoded, encoded_locations], axis=1)

# Display the preprocessed DataFrame
user_ratings_locations_encoded
```

```python
user_ratings_locations_encoded.info()
```

```python
from sklearn.cluster import KMeans

# Convert the 'ratings' column to numerical values
user_ratings_locations_encoded['ratings'] = user_ratings_locations_encoded['ratings'].apply(lambda x: [float(r) for r in x.split(', ')])
user_ratings_locations_encoded['ratings'] = user_ratings_locations_encoded['ratings'].apply(lambda x: sum(x) / len(x))

# Initialize the KMeans model with the desired number of clusters
kmeans = KMeans(n_clusters=5, random_state=42)

# Fit the model to the preprocessed data
kmeans.fit(user_ratings_locations_encoded)

# Add cluster labels to the DataFrame
user_ratings_locations_encoded['cluster'] = kmeans.labels_

# Display the DataFrame with cluster labels
print(user_ratings_locations_encoded.head())
```

```python
def recommend_destinations(user_id):
    # Get the cluster label of the user
    user_cluster = user_ratings_locations_encoded.loc[user_id, 'cluster']

    # Find other users in the same cluster
    users_in_cluster = user_ratings_locations_encoded[user_ratings_locations_encoded['cluster'] == user_cluster]

    # Get destinations visited by users in the same cluster
    destinations_visited = users_in_cluster.drop(columns=['ratings', 'cluster']).sum()

    # Remove destinations already visited by the user
    user_destinations = user_ratings_locations_encoded.loc[user_id, destinations_visited.index]
    recommendations = destinations_visited[destinations_visited.index != user_destinations]

    return recommendations[recommendations > 0].index.tolist()

# Test the function
user_id = 3  # Change this to the desired user ID
recommended_destinations = recommend_destinations(user_id)
print("Recommended destinations for user", user_id, ":", recommended_destinations)
```

```python
# Mapping of reviewer names to user IDs
name_to_id = {
    'John Smith': 1, 'Jane Doe': 2, 'David Lee': 3, 'Sarah Johnson': 4, 'Kim Nguyen': 5, 'Michael Brown': 6, 'Emily Davis': 7, 'Lucas Santos': 8, 'Laura Janssen': 9, 'Mohammed Ali': 10, 'Ana Hernandez': 11, 'Carlos Garcia': 12, 'Lily Wong': 13, 'Hans Mueller': 14, 'Fatima Khouri': 15, 'James MacKenzie': 16, 'Michael Chang': 17, 'Olivia Rodriguez': 18, 'Kenji Nakamura': 19, 'Emily Lee': 20, 'James Wilson': 21, 'Sofia Russo': 22, 'Raj Patel': 23, 'Lily Nguyen': 24, 'David Kim': 25, 'Maria Garcia': 26, 'Alice Smith': 27, 'Bob Johnson': 28, 'Charlie Lee': 29, 'Emma Davis': 30, 'Olivia Martin': 31, 'Harry Wilson': 32, 'Sophia Lee': 33, 'James Brown': 34, 'Mia Johnson': 35, 'William Davis': 36, 'Amelia Brown': 37, 'Adam Lee': 38, 'Sarah Wong': 39, 'Maria Silva': 40, 'Peter Brown': 41, 'Emma Garcia': 42, 'Michael Davis': 43, 'Nina Patel': 44, 'Kevin Kim': 45, 'Laura van den Berg': 46, 'Jennifer Nguyen': 47, 'Rachel Lee': 48, 'Jessica Wong': 49, 'Felipe Almeida': 50, 'Nisa Patel': 51, 'Ben Smith': 52, 'Laura Gomez': 53, 'Park Min Woo': 54, 'Michael Chen': 55, 'Sofia Rossi': 56, 'Rachel Sanders': 57, 'Emily Watson': 58, 'Ana Rodriguez': 59, 'Tom Wilson': 60, 'Olivia Green': 61, 'James Chen': 62, 'Lila Patel': 63, 'Marco Rossi': 64, 'Sarah Brown': 65, 'Sarah Lee': 66, 'Alex Kim': 67, 'Maria Hernandez': 68, 'Mark Johnson': 69, 'Amanda Chen': 70, 'Nana Kwon': 71, 'Tom Hanks': 72, 'Emma Watson': 73, 'Fatima Ahmed': 74, 'Liam Nguyen': 75, 'Giulia Rossi': 76, 'Putra Wijaya': 77, 'Kim Min-ji': 78, 'Emily Johnson': 79, 'Michael Wong': 80, 'Jessica Chen': 81, 'Ken Tanaka': 82, 'Rodrigo Oliveira': 83, 'Olivia Kim': 84, 'Robert Mueller': 85, 'Lisa Chen': 86, 'Emily Wong': 87, 'Mark Tan': 88, 'Emma Lee': 89, 'George Chen': 90, 'Sophia Kim': 91, 'Alex Ng': 92, 'Cindy Chen': 93, 'Emily Kim': 94, 'Frank Li': 95, 'Gina Lee': 96, 'Henry Kim': 97, 'Isabella Chen': 98, 'Jack Smith': 99, 'Katie Johnson': 100, 'John Doe': 101, 'Jane Smith': 102, 'Michael Johnson': 103, 'Jose Perez': 104, 'Emma Wilson': 105, 'Ryan Chen': 106, 'Sofia Rodriguez': 107, 'William Brown': 108
}

# Function to recommend common destinations among a list of user IDs
def recommend_common_destinations(user_ids):
    all_destinations = []  # List to store recommended destinations for all users
    for user_id in user_ids:
        recommended_destinations = recommend_destinations(user_id)  # Get recommended destinations for user
        all_destinations.extend(recommended_destinations)  # Add recommended destinations to the list

    # Count occurrences of each destination
    destination_counts = {destination: all_destinations.count(destination) for destination in set(all_destinations)}

    # Find destinations recommended for all users
    common_destinations = [destination for destination, count in destination_counts.items() if count == len(user_ids)]

    return common_destinations

# Function to recommend destinations for a given user ID
def recommend_destinations(user_id):
    # Get the cluster label of the user
    user_cluster = user_ratings_locations_encoded.loc[user_id, 'cluster']

    # Find other users in the same cluster
    users_in_cluster = user_ratings_locations_encoded[user_ratings_locations_encoded['cluster'] == user_cluster]

    # Get destinations visited by users in the same cluster
    destinations_visited = users_in_cluster.drop(columns=['ratings', 'cluster']).sum()

    # Remove destinations already visited by the user
    user_destinations = user_ratings_locations_encoded.loc[user_id, destinations_visited.index]
    recommendations = destinations_visited[destinations_visited.index != user_destinations]

    return recommendations[recommendations > 0].index.tolist()

# Function to recommend reviews for a given destination and budget
def recommend_reviews123(destination, budget):
    # Filter the DataFrame based on the user-entered destination and positive sentiment
    destination_reviews = df[(df['Destination'] == destination) & (df['Sentiment'] == 'Positive')]

    # Filter reviews based on budget
    destination_reviews = destination_reviews[destination_reviews['Total cost'] <= budget]

    # Sort the positive reviews by rating in descending order
    sorted_reviews = destination_reviews.sort_values(by='Rating', ascending=False)

    # Construct formatted recommendations
    recommendations = []
    reviewer_names = []  # List to store reviewer names
    for index, row in sorted_reviews.iterrows():
        recommendation = (
            f"Name: {row['Traveler name']}\n"
            f"Review: {row['review']}\n"
            f"Rating: {row['Rating']}\n"
            f"Cost Estimated: {row['Total cost']}\n"
        )
        recommendations.append(recommendation)
        reviewer_names.append(row['Traveler name'])  # Add reviewer name to the list

    return recommendations, reviewer_names

# User Interface
def user_interface123():
    global user_destination_global  # Declare the global variable
    user_ids_list = []  # List to store user IDs
    while True:
        user_destination = input("Enter the destination you want recommendations for (type 'exit' to quit): ")
        if user_destination.lower() == 'exit':
            break
        user_budget = int(input("Enter your budget: "))
        recommendations, reviewer_names = recommend_reviews123(user_destination, user_budget)

        if recommendations:
            print("\nRecommendations for {}: \n".format(user_destination))
            for recommendation in recommendations:
                print(recommendation)

            print("\nReviewers:")
            for name in reviewer_names:
                user_id = name_to_id.get(name)  # Get user ID from the dictionary
                print(f"{name} (User ID: {user_id})")
                user_ids_list.append(user_id)  # Add user ID to the list

        else:
            print("No positive reviews found within the specified budget for {}".format(user_destination))

        # Store the user destination in the global variable
        user_destination_global = user_destination

    return user_ids_list

# Run the user interface
user_ids = user_interface123()
print("User IDs:", user_ids)

# Get common destinations recommended for the users
common_destinations = recommend_common_destinations(user_ids)
# Filter out user's destination from common destinations
filtered_common_destinations = [dest for dest in common_destinations if dest != user_destination_global]

# Print recommendations excluding user's destination
print("\nCommon recommended destinations (excluding user's destination):")
for destination in filtered_common_destinations:
    print(destination)

# print("Common recommended destinations:", common_destinations)
```
